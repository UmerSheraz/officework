```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from homemade.linear_regression import LinearRegression
```

```python
data=pd.read_table('dat.txt' , sep='\t')

data.head(10)
```

```python
histohrams = data.hist(grid=True, figsize=(20,20))
```

```python
train_data= data.sample(frac=0.7)
test_data = data.drop(train_data.index)

input_param_name = 'Economy..GDP.per.Capita.'
output_param_name = 'Happiness.Score'

x_train= train_data[[input_param_name]].values
y_train= train_data[[output_param_name]].values

x_test= test_data[[input_param_name]].values
y_test= test_data[[output_param_name]].values
'''
plt.scatter(x_train, y_train, label='Training Dataset')
plt.scatter(x_test, y_test, label='test Dataset')
plt.xlabel(input_param_name)
plt.ylabel(output_param_name)
plt.title('C H')
plt.legend()
plt.show()
'''
```

```python
num_iterations = 500
regularization_param = 0
learning_rate=0.01
polymial_degree = 0
sinusoid_degree = 0

linear_regression =LinearRegression(x_train, y_train, polymial_degree,sinusoid_degree)


(theta, cost_history) = linear_regression.train(
    learning_rate,
    regularization_param,
    num_iterations
)
print('Initial cost: {:.2f}'.format(cost_history[5]))
print('Optimized cost: {:.2f}'.format(cost_history[-1]))
print(theta)

theta.flatten()

theta_table=pd.DataFrame({'Model parameters':theta.flatten()})
theta_table
```

```python
plt.plot(range(num_iterations), cost_history)
#plt.plot(range(100), cost_history[0:100])
plt.show()
```

```python
predictions_num = 100
x_predictions = np.linspace(x_train.min(),x_train.max(),predictions_num).reshape(predictions_num,1)
y_predictions = linear_regression.predict(x_predictions)
y_predictions

plt.scatter(x_train, y_train, label='Training Dataset')
plt.plot(x_train, y_train,'g', label='Training Dataset')
plt.scatter(x_test, y_test, label='test Dataset')
#plt.plot(x_test, y_test,'r', label='test Dataset')
plt.plot(x_predictions, y_predictions, 'b', label='Prediction')
plt.scatter(x_predictions, y_predictions, label='Prediction')
plt.xlabel(input_param_name)
plt.ylabel(output_param_name)
plt.title('C H')
plt.legend()
plt.show()
```
